```python
from matplotlib import markers
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.lines import Line2D

df_precision = pd.read_csv("Precision.csv")

df_precision_melted = pd.melt(df_precision, id_vars=['Report'], var_name='Method', value_name='Precision')
plt.figure(figsize=(9, 5))
sns.boxplot(x='Precision', y='Method', data=df_precision_melted)
sns.swarmplot(x='Precision', y='Method', data=df_precision_melted, color=".25")

legend_elements = [Line2D([0], [0], marker='o', color='none', label='One report',
                          markerfacecolor='black', markersize=4),
                   Line2D([0], [0], marker='D', color='none', label='Outlier',
                          markerfacecolor='black', markersize=4)
                 ]
#  Line2D([0], [0], marker='D', color='none', label='Two reports',
#                           markerfacecolor='black', markersize=4)
plt.legend(handles=legend_elements, loc='best')


plt.xticks(np.arange(0, 1.1, step=0.1), rotation=90)
plt.title('Precision of Different Methods')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the file
df = pd.read_csv('Sentence_comparison.csv', sep=', ')

# Create a bar chart for the second and third columns
df.plot(x='Report', y=['Number of sentences (original)', 'Number of sentences (aCTIon approach)'], kind='bar')

plt.title('Number of Sentences Comparison')
plt.ylabel('Number of Sentences')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
